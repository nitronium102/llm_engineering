and responds with an explanation. This is a tool that you will be able to use yourself during the course!

````python
# imports
from IPython.display import Markdown, display, update_display
from openai import OpenAI
import ollama
````

````python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
````

````python
# set up environment
OLLAMA_BASE_URL = "http://localhost:11434/v1"
openai = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
````

````python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
````

````python
system_prompt = """
    You are a helpful technical tutor who answers questions about python code, 
    software engineering, machine learning, and LLMs.
    You are a senior software engineer with 20 years of experience in the field.
    You are also an expert in teaching and explaining complex ideas.
    Now you have to give a precise and detailed explanation to your user's question.
"""
user_prompt = input("Enter your question: ")

# user_prompt = """
#     Please explain a detailed explanation to the following question.
# """ + question

````

````python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
````

````python
# Get gpt-4o-mini to answer, with streaming
````

````python
# Get Llama 3.2 to answer
# 토큰이 생성되는 즉시 stream으로 받음 (아직 응답 미완성)
stream = ollama.chat(
    model=MODEL_LLAMA,
    messages=messages,
    stream=True
)

# 누적 출력용 버퍼
response = ""
# Jupiter에서 같은 영역을 계속 갱신하기 위한 handle (한 화면에서 타이핑처럼 보이게)
display_handle = display(Markdown(""), display_id=True)

for chunk in stream:
    # Ollama ChatResponse 구조
    if hasattr(chunk, "message") and chunk.message and chunk.message.content:
        response += chunk.message.content

    # 2) 렌더링 직전에 최소한의 후처리
    cleaned = response.replace("```markdown", "").replace("```", "")
    update_display(Markdown(cleaned), display_id=display_handle.display_id)


````
